## 1. 회귀 분석(Regression)


 ### - training data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris = datasets.load_iris()      #data과 target 으로 나눠져 있음.

x_original = iris.data

x = x_original[:,0]     # 꽃받침의 길이
y = x_original[:,2]     # 꽃잎의 길이

plt.scatter(x,y,c='b')
plt.xlabel('sepal lenght')
plt.ylabel('petal lenght')

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3)    #test set과 train set을 나눠주는 역할


from sklearn.linear_model import LinearRegression
lig = LinearRegression()                 # model 생성
lig.fit(x_train.reshape(-1,1),y_train)   # model 학습 (reshape은 하나의 열로 표현해주는 것)
lig.coef_, lig.intercept_

w = lig.coef_
b = lig.intercept_

xx = np.linspace(4,8,10)
yy = w*xx +b

plt.xlabel('sepal lenght')
plt.ylabel('petal lenght')
plt.plot(xx, yy, c='r')        # 회귀선
plt.scatter(x_train, y_train, c='r' ) # training data의 산점도
plt.scatter(x_test, y_test, c='b' ) # test data의 산점도


   
```

## 2. 분류 분석(Classification)


 ### - training data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

iris=datasets.load_iris()      #data과 target 으로 나눠져 있음.

x_original = iris.data
y = iris.target

x2 = x_original[:,[0,3]]     # extract only the first tow columns
y2 = y.copy()
y2[y2==2]=1     # y2==2 인 경우, y2==1 로 변경한다.

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x2,y2,test_size=0.3, random_state=7)   #test set과 train set을 나눠주는 역할

from sklearn.linear_model import SGDClassifier    #머신러닝을 위한 파트
clf = SGDClassifier()                             # model 생성
clf.fit(x_train,y_train)                          # model 학습
clf.coef_, clf.intercept_
```

  ### - test data

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = clf.predict(x_test)
accuracy_score(y_test, y_pred)

# train data
plt.figure(figsize=(8,6))
my_marker = ['*','o']
my_colors = ['r','b']

for i in range(2) :
    xs = x_train[:,0][y_train==i]
    ys = x_train[:,1][y_train==i]
    plt.scatter(xs,ys,marker=my_marker[i],c=my_colors[i])


plt.xlabel('sepal lenght')
plt.ylabel('sepal width')

# train_classification
a = clf.coef_[0,0]
b = clf.coef_[0,1]
c = clf.intercept_[0]
xx = np.linspace(4,8,100)
yy = - a/b*xx - c/b

plt.plot(xx,yy)

# test data
my_marker2 = ['v','^']
my_colors2 = ['g','k']

for i in range(2) :
    xs2 = x_test[:,0][y_test==i]
    ys2 = x_test[:,1][y_test==i]
    plt.scatter(xs2,ys2,marker=my_marker2[i],c=my_colors2[i])

names = ['train_regression','train_setosa','train_non-setosa','test_setosa','test_non-setosa']
plt.legend(names)
```
