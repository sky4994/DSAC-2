```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import font_manager, rc
path = 'c:\Windows\Fonts\malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False

from sklearn import datasets

iris=datasets.load_iris()      #data과 target 으로 나눠져 있음.

x = iris.data
y = iris.target

x2 = x[:,:2]     # extract only the first tow columns
```

```python
plt.figure(figsize=(8,6))
my_marker = ['*','o','v']
my_colors = ['r','b','g']

for i in range(3) :
    xs = x2[:,0][y==i]
    ys = x2[:,1][y==i]
    plt.scatter(xs,ys,marker=my_marker[i],c=my_colors[i])

plt.title('Viewing the dataset', fontsize=25)
plt.legend(iris.target_names)
plt.xlabel('sepal lenght')
plt.ylabel('sepal width')
```

```python
y2 = y.copy()
y2[y2==2]=1     # y2==2 인 경우, y2==1 로 변경한다.
```

```python
plt.figure(figsize=(8,6))
my_marker = ['*','o']
my_colors = ['r','b']

for i in range(2) :
    xs = x2[:,0][y==i]
    ys = x2[:,1][y==i]
    plt.scatter(xs,ys,marker=my_marker[i],c=my_colors[i])

plt.title('Viewing the dataset: setosa & non-setosa', fontsize=25)
names = ['setosa','non=setosa']
plt.legend(names)
plt.xlabel('sepal lenght')
plt.ylabel('sepal width')
```

```python
from sklearn.linear_model import SGDClassifier    #머신러닝을 위한 파트
model= SGDClassifier()
model.fit(x2,y2)
model.coef_, model.intercept_
```

```python
#임의로 만든 data로 잘 구동되나 실험해보기 

np.random.seed(42)
z = np.random.rand(10,2).round(2)*10
z, model.predict(z)
```

```python
#non Linear classifier using decisiontree

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib import font_manager, rc
path = 'c:\Windows\Fonts\malgun.ttf'
font_name = font_manager.FontProperties(fname=path).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False

from sklearn import datasets

iris=datasets.load_iris()      #data과 target 으로 나눠져 있음.

x = iris.data
y = iris.target

x2 = x[:,:2]     # extract only the first tow columns
y2 = y.copy()
y2[y2==2]=1     # y2==2 인 경우, y2==1 로 변경한다.


from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x2,y2)
dt_model.classes_
```

```python
#임의로 만든 data로 잘 구동되나 실험해보기 

np.random.seed(42)
z = np.random.rand(10,2).round(2)*10
z, dt_model.predict(z)
```

이렇게 어떤 모델을 쓰느냐에 따라 조건이 같아도 다르게 구분하게 된다.
